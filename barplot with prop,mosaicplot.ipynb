```python
import pandas as pd
df=pd.read_excel("C:\\resources\\resources\\mean median mode.xlsx")
pd.crosstab(df.department,df.gender,normalize=True)
```

```python
#row wise proportions
pd.crosstab(df.department,df.gender,normalize='index')
```

```python
#column wise proportions
pd.crosstab(df.department,df.gender,normalize='columns')
```

```python
#barchart with proportions
pd.crosstab(df.department,df.gender).plot.bar(stacked=True)
```

```python
#genderwise barchart
pd.crosstab(df.gender,df.department).plot.bar(stacked=True)
```

```python
#gender wise proportionable of employee
pd.crosstab(df.gender,df.department,normalize='index').plot.bar(stacked=True)
```

```python
#department wise proportionable of employee
pd.crosstab(df.department,df.gender,normalize='index').plot.bar(stacked=True)
```
